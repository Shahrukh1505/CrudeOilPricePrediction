```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://pkgstore.datahub.io/core/oil-prices/wti-day_csv/data/3ee9614b39c99ec7ec5e35c3df3deff3/wti-day_csv.csv")
```

```python
plt.figure(figsize=(15,10))
df['WTI Spot Price'].plot(kind='bar')
plt.title('Variation of oil price over years')
plt.ylabel('Price')
plt.xticks(color = 'w',rotation=90)
```

```python
plt.figure(figsize=(15,10))
df['WTI Spot Price'].plot(kind='line')
plt.title('Variation of oil price over years')
plt.ylabel('Price')
plt.xticks(color = 'w')
```

```python
l = []
for i in df['Date']:
    l.append(int(i[0]+i[1]+i[2]+i[3]))
df['year'] = l;
```

```python
df.head()
```

```python
x=df['year']
y=df['WTI Spot Price']
plt.bar(x,y)
plt.title('Variation of Price over years')
plt.xlabel('Year')
plt.ylabel('Price in $')
```
