```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Bidirectional

# Preprocess the dataset
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df['Price'].values.reshape(-1, 1))

# Split the data into train and test sets
train_size = int(len(df) * 0.8)
train_data = scaled_data[:train_size]
test_data = scaled_data[train_size:]

# Function to create the time series dataset
def create_time_series_dataset(data, lookback):
    X, Y = [], []
    for i in range(len(data) - lookback):
        X.append(data[i:(i + lookback), 0])
        Y.append(data[i + lookback, 0])
    return np.array(X), np.array(Y)

# Define the lookback period
lookback = 10

# Create the time series training dataset
X_train, Y_train = create_time_series_dataset(train_data, lookback)

# Create the time series testing dataset
X_test, Y_test = create_time_series_dataset(test_data, lookback)

# Reshape the input data for LSTM
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

# Build the LSTM model
lstm_model = Sequential()
lstm_model.add(LSTM(50, input_shape=(lookback, 1)))
lstm_model.add(Dense(1))
lstm_model.compile(optimizer='adam', loss='mean_squared_error')

# Fit the LSTM model
lstm_model.fit(X_train, Y_train, epochs=100, batch_size=32, verbose=1)

# Build the Bi-LSTM model
bilstm_model = Sequential()
bilstm_model.add(Bidirectional(LSTM(50, return_sequences=True), input_shape=(lookback, 1)))
bilstm_model.add(Bidirectional(LSTM(50)))
bilstm_model.add(Dense(1))
bilstm_model.compile(optimizer='adam', loss='mean_squared_error')

# Fit the Bi-LSTM model
bilstm_model.fit(X_train, Y_train, epochs=100, batch_size=32, verbose=1)

# Evaluate the models
train_predictions_lstm = lstm_model.predict(X_train)
test_predictions_lstm = lstm_model.predict(X_test)

train_predictions_bilstm = bilstm_model.predict(X_train)
test_predictions_bilstm = bilstm_model.predict(X_test)

# Inverse transform the predictions
train_predictions_lstm = scaler.inverse_transform(train_predictions_lstm)
test_predictions_lstm = scaler.inverse_transform(test_predictions_lstm)

train_predictions_bilstm = scaler.inverse_transform(train_predictions_bilstm)
test_predictions_bilstm = scaler.inverse_transform(test_predictions_bilstm)

# Inverse transform the actual values
Y_train = scaler.inverse_transform([Y_train])
Y_test = scaler.inverse_transform([Y_test])

# Calculate the root mean squared error (RMSE)
train_rmse_lstm = np.sqrt(np.mean((train_predictions_lstm - Y_train) ** 2))
test_rmse_lstm = np.sqrt(np.mean((test_predictions_lstm - Y_test) ** 2))

train_rmse_bilstm = np.sqrt(np.mean((train_predictions_bilstm - Y_train) ** 2))
test_rmse_bilstm = np.sqrt(np.mean((test_predictions_bilstm - Y_test) ** 2))

print("LSTM - Train RMSE:", train_rmse_lstm)
print("LSTM - Test RMSE:", test_rmse_lstm)

print("Bi-LSTM - Train RMSE:", train_rmse_bilstm)
print("Bi-LSTM - Test RMSE:", test_rmse_bilstm)
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Bidirectional

# Load the dataset
dataset_path = r"C:\Users\Shahrukh Ansari\Desktop\majorProject\files\russia.csv"
df = pd.read_csv(dataset_path)

# Preprocess the dataset
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df['Price'].values.reshape(-1, 1))

# Split the data into train and test sets
train_size = int(len(df) * 0.8)
train_data = scaled_data[:train_size]
test_data = scaled_data[train_size:]

# Function to create the time series dataset
def create_time_series_dataset(data, lookback):
    X, Y = [], []
    for i in range(len(data) - lookback):
        X.append(data[i:(i + lookback), 0])
        Y.append(data[i + lookback, 0])
    return np.array(X), np.array(Y)

# Define the lookback period
lookback = 10

# Create the time series training dataset
X_train, Y_train = create_time_series_dataset(train_data, lookback)

# Create the time series testing dataset
X_test, Y_test = create_time_series_dataset(test_data, lookback)

# Reshape the input data for LSTM
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

# Build the LSTM model
lstm_model = Sequential()
lstm_model.add(LSTM(50, input_shape=(lookback, 1)))
lstm_model.add(Dense(1))
lstm_model.compile(optimizer='adam', loss='mean_squared_error')

# Fit the LSTM model
lstm_model.fit(X_train, Y_train, epochs=100, batch_size=32, verbose=1)

# Build the Bi-LSTM model
bilstm_model = Sequential()
bilstm_model.add(Bidirectional(LSTM(50, return_sequences=True), input_shape=(lookback, 1)))
bilstm_model.add(Bidirectional(LSTM(50)))
bilstm_model.add(Dense(1))
bilstm_model.compile(optimizer='adam', loss='mean_squared_error')

# Fit the Bi-LSTM model
bilstm_model.fit(X_train, Y_train, epochs=100, batch_size=32, verbose=1)

# Evaluate the models
train_predictions_lstm = lstm_model.predict(X_train)
test_predictions_lstm = lstm_model.predict(X_test)

train_predictions_bilstm = bilstm_model.predict(X_train)
test_predictions_bilstm = bilstm_model.predict(X_test)

# Inverse transform the predictions
train_predictions_lstm = scaler.inverse_transform(train_predictions_lstm)
test_predictions_lstm = scaler.inverse_transform(test_predictions_lstm)

train_predictions_bilstm = scaler.inverse_transform(train_predictions_bilstm)
test_predictions_bilstm = scaler.inverse_transform(test_predictions_bilstm)

# Inverse transform the actual values
Y_train = scaler.inverse_transform([Y_train])
Y_test = scaler.inverse_transform([Y_test])

# Calculate performance metrics
train_rmse_lstm = np.sqrt(mean_squared_error(Y_train[0], train_predictions_lstm[:, 0]))
test_rmse_lstm = np.sqrt(mean_squared_error(Y_test[0], test_predictions_lstm[:, 0]))

train_mae_lstm = mean_absolute_error(Y_train[0], train_predictions_lstm[:, 0])
test_mae_lstm = mean_absolute_error(Y_test[0], test_predictions_lstm[:, 0])

train_mse_lstm = mean_squared_error(Y_train[0], train_predictions_lstm[:, 0])
test_mse_lstm = mean_squared_error(Y_test[0], test_predictions_lstm[:, 0])

train_rmse_bilstm = np.sqrt(mean_squared_error(Y_train[0], train_predictions_bilstm[:, 0]))
test_rmse_bilstm = np.sqrt(mean_squared_error(Y_test[0], test_predictions_bilstm[:, 0]))

train_mae_bilstm = mean_absolute_error(Y_train[0], train_predictions_bilstm[:, 0])
test_mae_bilstm = mean_absolute_error(Y_test[0], test_predictions_bilstm[:, 0])

train_mse_bilstm = mean_squared_error(Y_train[0], train_predictions_bilstm[:, 0])
test_mse_bilstm = mean_squared_error(Y_test[0], test_predictions_bilstm[:, 0])

print("LSTM - Train RMSE:", train_rmse_lstm)
print("LSTM - Test RMSE:", test_rmse_lstm)
print("LSTM - Train MAE:", train_mae_lstm)
print("LSTM - Test MAE:", test_mae_lstm)
print("LSTM - Train MSE:", train_mse_lstm)
print("LSTM - Test MSE:", test_mse_lstm)

print("Bi-LSTM - Train RMSE:", train_rmse_bilstm)
print("Bi-LSTM - Test RMSE:", test_rmse_bilstm)
print("Bi-LSTM - Train MAE:", train_mae_bilstm)
print("Bi-LSTM - Test MAE:", test_mae_bilstm)
print("Bi-LSTM - Train MSE:", train_mse_bilstm)
print("Bi-LSTM - Test MSE:", test_mse_bilstm)
```
