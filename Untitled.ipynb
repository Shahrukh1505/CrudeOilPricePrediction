```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
from math import sqrt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import tensorflow as tf

# Preprocess the dataset
data = df['Price'].values.reshape(-1, 1)  # Extract the 'Price' column
data = np.nan_to_num(data)

scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)

# Split the dataset into training and testing sets
train_data_len = int(0.8 * len(scaled_data))
train_data = scaled_data[:train_data_len]
test_data = scaled_data[train_data_len:]

# Prepare the data for LSTM
def create_dataset(dataset, lookback):
    X, y = [], []
    for i in range(len(dataset) - lookback):
        X.append(dataset[i:i + lookback, 0])
        y.append(dataset[i + lookback, 0])
    return np.array(X), np.array(y)

lookback = 30
X_train, y_train = create_dataset(train_data, lookback)
X_test, y_test = create_dataset(test_data, lookback)

X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

# Build and train the LSTM model
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(LSTM(units=50))
model.add(Dense(units=1))

model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(X_train, y_train, epochs=10, batch_size=32)

# Make predictions using the trained model
predictions = model.predict(X_test)

# Reverse the scaling of predictions
predictions = scaler.inverse_transform(predictions)
y_test = scaler.inverse_transform(y_test.reshape(-1, 1))

# Evaluation metrics
mae = mean_absolute_error(y_test, predictions)
mse = mean_squared_error(y_test, predictions)
rmse = sqrt(mse)

print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.svm import SVR
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Bidirectional

# Load the dataset
dataset_path = r"C:\Users\Shahrukh Ansari\Desktop\majorProject\files\majorFirst.csv"
data = pd.read_csv(dataset_path)

# Preprocess the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data['Price'].values.reshape(-1, 1))

# Set the time steps for input sequence
time_steps = 60

# Create the input and output sequences
X, y = [], []
for i in range(len(scaled_data) - time_steps - 1):
    a = scaled_data[i:(i + time_steps), 0]
    X.append(a)
    y.append(scaled_data[i + time_steps, 0])
X, y = np.array(X), np.array(y)

# Reshape the input data for LSTM (samples, time steps, features)
X_lstm = np.reshape(X, (X.shape[0], X.shape[1], 1))

# Define the number of folds for cross-validation
k = 5

# Perform k-fold cross-validation
kf = KFold(n_splits=k, shuffle=True)
fold = 1

linear_reg_losses = []
rf_losses = []
xgb_losses = []
svr_losses = []
lstm_losses = []
bi_lstm_losses = []

for train_index, test_index in kf.split(X):
    print(f"Fold: {fold}")
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    X_lstm_train, X_lstm_test = X_lstm[train_index], X_lstm[test_index]

    # Linear Regression
    linear_reg = LinearRegression()
    linear_reg.fit(X_train, y_train)
    linear_reg_loss = np.mean((linear_reg.predict(X_test) - y_test) ** 2)
    linear_reg_losses.append(linear_reg_loss)

    # Random Forest Regressor
    rf = RandomForestRegressor()
    rf.fit(X_train, y_train)
    rf_loss = np.mean((rf.predict(X_test) - y_test) ** 2)
    rf_losses.append(rf_loss)

    # XGBoost Regressor
    xgb = XGBRegressor()
    xgb.fit(X_train, y_train)
    xgb_loss = np.mean((xgb.predict(X_test) - y_test) ** 2)
    xgb_losses.append(xgb_loss)

    # Support Vector Regressor
    svr = SVR()
    svr.fit(X_train, y_train)
    svr_loss = np.mean((svr.predict(X_test) - y_test) ** 2)
    svr_losses.append(svr_loss)

    # LSTM
    lstm_model = Sequential()
    lstm_model.add(LSTM(64, input_shape=(X_lstm_train.shape[1], 1)))
    lstm_model.add(Dense(1))
    lstm_model.compile(loss='mean_squared_error', optimizer='adam')
    lstm_model.fit(X_lstm_train, y_train, epochs=10, batch_size=32, verbose=0)
    lstm_loss = lstm_model.evaluate(X_lstm_test, y_test, verbose=0)
    lstm_losses.append(lstm_loss)

    # Bi-LSTM
    bi_lstm_model = Sequential()
    bi_lstm_model.add(Bidirectional(LSTM(64, return_sequences=True), input_shape=(X_lstm_train.shape[1], 1)))
    bi_lstm_model.add(Bidirectional(LSTM(64)))
    bi_lstm_model.add(Dense(1))
    bi_lstm_model.compile(loss='mean_squared_error', optimizer='adam')
    bi_lstm_model.fit(X_lstm_train, y_train, epochs=10, batch_size=32, verbose=0)
    bi_lstm_loss = bi_lstm_model.evaluate(X_lstm_test, y_test, verbose=0)
    bi_lstm_losses.append(bi_lstm_loss)

    fold += 1

# Calculate and print the average loss for each model
print("Linear Regression Average Loss:", np.mean(linear_reg_losses))
print("Random Forest Regressor Average Loss:", np.mean(rf_losses))
print("XGBoost Regressor Average Loss:", np.mean(xgb_losses))
print("Support Vector Regressor Average Loss:", np.mean(svr_losses))
print("LSTM Average Loss:", np.mean(lstm_losses))
print("Bi-LSTM Average Loss:", np.mean(bi_lstm_losses))


```
