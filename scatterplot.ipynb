# Scatter Plot

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]
plt.scatter(x, y, color="r")
plt.title("Scatter Plot", fontsize=20)
plt.xlabel("Day")
plt.ylabel("No.")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]
colors = ["r","y","g","b","r","g","r"]
sizes = [100,20,40,50,20,10,70]
plt.scatter(x, y, color=colors, s=sizes,alpha=1, marker="*",edgecolor="black",linewidth=2)
plt.title("Scatter Plot", fontsize=20)
plt.xlabel("Day")
plt.ylabel("No.")
plt.show()

#0-100  ColorMaps
```
